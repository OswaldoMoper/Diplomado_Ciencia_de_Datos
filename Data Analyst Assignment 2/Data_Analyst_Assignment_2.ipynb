### Table 1 : Employee Information

identification, job titles, management hierarchies, tenure, and work region

### Table 3: Performance Data

This table includes the sales and revenue generation activities of employees by cataloguing
opportunities, their progression, and financial outcomes. It captures granular data on sales
stage milestones, product-related charges, and revenue figures, all linked by employee and
opportunity identifiers, making it a vital asset for analyzing sales performance and compensation

* **Product Rate Plan Charge:** The charge associated with the product's rate plan.

### Question 2

2. How would you analyze the performance of an account executive? Is there a difference
between those who have completed training and those who have not? How would you
segment the data and present your findings to senior stakeholders?
* Hint: Look at this from an overall employee perspective who has completed any
training vs. those who have not completed any training. Any training would count,

rather than distinguishing which specific courses were completed, as they all
contribute to one Suite product.


---

#### Code

```python
Per1 = PerD.groupby( [ 'Employee_ID', 'Add-On ARR (converted) Currency', 'Total Commissionable ARR (converted) Currency' ] ).agg( Total_Add_On_ARR = ( 'Add-On ARR (converted)', sum ), Sum_Total_Comissionable = ( 'Total Commissionable ARR (converted)', sum )  ).reset_index()
Per1
```

```python
TC = ComT.groupby ( 'Employee_ID' ).count().reset_index()
TC
```

```python
TC1 = TC.merge( Per1, on = [ 'Employee_ID' ], how = 'outer'  )
TC1.Training_Completed = TC1.Training_Completed.fillna('0').astype(int)
TC1.Total_Add_On_ARR = TC1.Total_Add_On_ARR.fillna('0').astype(int)
TC1.Sum_Total_Comissionable = TC1.Sum_Total_Comissionable.fillna('0').astype(int)
TC1
```

```python
TC2 = TC1.drop( columns = [ 'Employee_ID' ] )
TC2.info()
```

#### Results

```python
Tcorr = TC2.corr( numeric_only = True )
Tcorr
```

```python
Tcov = TC2.cov( numeric_only = True )
Tcov
```

####Graphics

```python
TC3 = TC2.copy()
TC_type = CategoricalDtype(categories = ['0', '1', '2', '3', '4', '5'] , ordered = True)
TC3.Training_Completed = TC3.Training_Completed.astype(str)
TC3.Training_Completed = TC3.Training_Completed.astype(TC_type)
TC3 = TC3.sort_values( by ='Training_Completed' )
```

```python
CAdd = np.polyfit( TC2.Training_Completed, TC2.Total_Add_On_ARR, 1 )
AT1 = np.poly1d(CAdd)

Ccom = np.polyfit( TC2.Training_Completed, TC2.Sum_Total_Comissionable, 1 )
CT1 = np.poly1d(Ccom)
```

```python
STC = px.scatter(
      TC3
    , x = 'Training_Completed'
    , y = 'Total_Add_On_ARR'
    )
trendAdd = px.line (
      TC2
    , x = 'Training_Completed'
    , y = AT1(TC2.Training_Completed)
    , color_discrete_sequence= [ "#ab1111" , "#940f0f" ]
    # , mode = 'lines'
    # , name = 'Trend Line'
)
STC.add_trace(trendAdd.data[0])
STC.update_layout(
      yaxis_title = 'Add-On ARR'
    , xaxis_title = 'Completed Trainings'
    , template = 'plotly_dark'
    , width = 1600
    , height = 900
)
STC.show()
```

```python
STC1 = px.scatter(
      TC3
    , x = 'Training_Completed'
    , y = 'Sum_Total_Comissionable')
trendCom = px.line (
      TC2
    , x = 'Training_Completed'
    , y = CT1(TC2.Training_Completed)
    , color_discrete_sequence= [ "#ab1111" , "#940f0f" ]
    # , mode = 'lines'
    # , name = 'Trend Line'
)
STC1.add_trace(trendCom.data[0])
STC1.update_layout(
      yaxis_title = 'Total Comissionable'
    , xaxis_title = 'Completed Trainings'
    , template = 'plotly_dark'
    , width = 1600
    , height = 900
)
STC1.show()
```

#### Conclusion

```python
print(Tcorr.to_string())
```

The graphs with a trend line indicates that performance improves with the completion of just one training.

And considering the correlation coefficients, it can be inferred that the number of training sessions has a negative impact on performance. This holds true when we talk about more than one training session.

### Question 3

Any other interesting insights that you can see in the data?

**R=** We have previously observed the relationship between training completion and employee performance. In this case, the segments by leaders can be evaluated by examining the number of opportunities, the total commissions, or the average commissions per leader.

As we will do next:

#### Code

```python
Per2 = PerD.groupby( [ 'Employee_ID', 'Opportunity ID'] ).agg( Sum_Total_Comissionable = ( 'Total Commissionable ARR (converted)', sum )  ).reset_index()
```

```python
ComSVP = Per2.groupby( [ 'Employee_ID' ] ).agg ( Sum_Total_Comissionable = ( 'Sum_Total_Comissionable', sum ) )
ComSVP[ 'Opportunities' ] =  Per2.groupby( [ 'Employee_ID' ] )['Opportunity ID'].count()
ComSVP[ 'Mean' ] =  ComSVP.Sum_Total_Comissionable / ComSVP.Opportunities
ComSVP[ 'Mean' ] = ComSVP.Mean.astype (int)
```

```python
CombyL = ComSVP.merge( A2, on = [ 'Employee_ID' ], how = 'left'  ).groupby( [ 'SVP Leader' ] ).agg( Opportunities = ( 'Opportunities', sum ), Sum_Total_Comissionable = ( 'Sum_Total_Comissionable', sum ) ).reset_index()
CombyL[ 'Mean' ] = CombyL.Sum_Total_Comissionable / CombyL.Opportunities
CombyL[ 'Mean' ] = CombyL.Mean.astype(int)
```

#### Results

```python
print('Table 1. Sort by Opportunities - descending\n')
print(CombyL.sort_values( by = 'Opportunities', ascending = False ).to_string())
```

```python
print('Table 2. Sort by Sum_Total_Comissionable - descending\n')
print(CombyL.sort_values( by = 'Sum_Total_Comissionable', ascending = False ).to_string())
```

```python
print('Table 3. Sort by Mean - descending\n')
print(CombyL.sort_values( by = 'Mean', ascending = False ).to_string())
```

In Table 1, we can see that the Leader 3 sector is the one that compares the greatest number of opportunities, while Leader 10 is the poorest in terms of number of opportunities.

In the table 2, we can see that the Leader 1 sector is the one that obtains the highest amount of commissions, while the Leader 13 sector is the one that obtains the lowest amount of commissions.

And finally, in Table 3 it is notable that the Leader 10 sector is the one that has the highest average of commissions despite having the least number of opportunities, contrasting with that of Leader 3 having the lowest average of commissions, with the highest number of opportunities.

####Graphics

```python
#Función para dar formato al titulo
def format_title(title, subtitle=None, subtitle_font_size=14):
        title = f'<b>{title}</b>'
        if not subtitle:
            return title
        subtitle = f'<span style="font-size: {subtitle_font_size}px;">{subtitle}</span>'
        return f'{title}<br>{subtitle}'
```

```python
CL_type = CategoricalDtype(categories = ['Leader 1', 'Leader 2', 'Leader 3', 'Leader 4', 'Leader 5', 'Leader 6', 'Leader 7', 'Leader 8', 'Leader 9', 'Leader 10', 'Leader 11', 'Leader 12', 'Leader 13', 'Leader 14' ] , ordered = True)
CombyL[ 'SVP Leader' ] = CombyL['SVP Leader'].astype(CL_type)
CombyL = CombyL.sort_values( by = 'SVP Leader')
CombyL
```

```python
SVP1 = px.bar(
       CombyL
    , x = 'SVP Leader'
    , y = 'Opportunities'
    , color = 'Sum_Total_Comissionable'
    )
SVP1.update_layout(
      xaxis_title = 'SVP Leader'
    , yaxis_title = 'Opportunities'
    , template = 'plotly_dark'
    , title = format_title('Opportunities per SVP Leader', 'Colored by Sum of Comissions')
    , width = 1600
    , height = 900
)
SVP1.show()
```

```python
SVP2 = px.bar(
      CombyL
    , x = 'SVP Leader'
    , y = 'Mean'
    , color = 'Sum_Total_Comissionable'
    )
SVP2.update_layout(
      xaxis_title = 'SVP Leader'
    , yaxis_title = 'Mean of comissions'
    , template = 'plotly_dark'
    , title = format_title ('Mean of comissions per SVP Leader', 'Colored by Sum of Comissions')
    , width = 1600
    , height = 900
)
SVP2.show()
```

After reviewing the colored graphs, it is clear that:

*   The sector linked to Leader 4 is consistent in the three categories we use to evaluate the performance of sectors.
*   The strength of the sector related to Leader 1 is the amount of commissions, which is reflected in its total commissions, although it loses a little being one of the poorest in the average commissions per opportunity.
*   Even though the Leader 3 sector is the one with the greatest number of opportunities, the fact that it's the sector with the lowest average hurts it so much that it also falters in the category of total commissions.
*   Despite a lower overall performance than the Leader 4 sector, the sector related to Leader 11 is still consistent.
*   Despite having the best average commissions, the Leader 10 sector loses many points in total commissions because it has the fewest opportunities.
*   In general, the other sectors can follow the Leader 4 sector's footsteps, which would lead to consistent performance.
