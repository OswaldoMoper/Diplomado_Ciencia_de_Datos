```python
#Función para dar formato al titulo
def format_title(title, subtitle=None, subtitle_font_size=14):
        title = f'<b>{title}</b>'
        if not subtitle:
            return title
        subtitle = f'<span style="font-size: {subtitle_font_size}px;">{subtitle}</span>'
        return f'{title}<br>{subtitle}'
```

```python
fig = px.bar(df_exl , x = 'experience_level', y = 'Count')
fig.update_layout(
    yaxis_title = 'Conteo',
    xaxis_title = 'Nivel de Experiencia',
    title_text = format_title('Gráfca de Barras','Clase Diplomado'),
    template = 'plotly_dark'
)
fig
```

#Multi-Graphics

```python
df_rrn['work_year'] = df_rrn.work_year.astype(int)
```

```python
df_sal = pd.DataFrame(df.groupby(['work_year', 'remote_ratio']).salary_in_usd.mean()).reset_index()
df_sal
```

```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots
```

```python
figm = make_subplots(specs = [ [ { 'secondary_y':True } ] ] )
figm.add_trace( go.Bar( x = df_rrn.work_year, y = df_rrn['count'] ) )
```

```python
figm = make_subplots(specs = [ [ { 'secondary_y':True } ] ] )
for ratio in df_rrn.remote_ratio.unique():
  figm.add_trace( go.Bar( x = df_rrn[ df_rrn.remote_ratio == ratio ].work_year, y = df_rrn[ df_rrn.remote_ratio == ratio ]['count'], name = str(ratio) ) )
figm
```

```python
figm = make_subplots(specs = [ [ { 'secondary_y':True } ] ] )
for ratio in df_rrn.remote_ratio.unique():
  figm.add_trace( go.Bar( x = df_rrn[ df_rrn.remote_ratio == ratio ].work_year, y = df_rrn[ df_rrn.remote_ratio == ratio ]['count'], name = str(ratio) ) )
figm.add_trace( go.Scatter( x = df_sal.work_year, y = df_sal.salary_in_usd, name = str(ratio) ), secondary_y = True )
figm
```

```python
figm = make_subplots(specs = [ [ { 'secondary_y':True } ] ] )
for ratio in df_rrn.remote_ratio.unique():
  figm.add_trace( go.Bar( x = df_rrn[ df_rrn.remote_ratio == ratio ].work_year, y = df_rrn[ df_rrn.remote_ratio == ratio ]['count'], name = str(ratio) ) )
for ratio in df_sal.remote_ratio.unique():
  figm.add_trace( go.Scatter( x = df_sal[ df_sal.remote_ratio == ratio ].work_year, y = df_sal[ df_sal.remote_ratio == ratio ].salary_in_usd, name = str(ratio) ), secondary_y = True )
figm
```
